```python
import math

def plot_all_histograms(df, title_prefix=""):
    num_cols = df.select_dtypes(include=[np.number]).columns
    n_cols = 3
    n_rows = math.ceil(len(num_cols) / n_cols)

    plt.figure(figsize=(5 * n_cols, 4 * n_rows))

    for i, col in enumerate(num_cols, 1):
        plt.subplot(n_rows, n_cols, i)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f"{title_prefix}{col}")
        plt.xlabel("")
        plt.ylabel("")

    plt.tight_layout()
    plt.show()
```

```python
plot_all_histograms(df)
```

```python
from sklearn.cluster import KMeans, HDBSCAN, AgglomerativeClustering
kmeans = KMeans(n_clusters=2)
hc = AgglomerativeClustering(n_clusters=2)
hdb = HDBSCAN(min_cluster_size=7, min_samples=3)
```

```python
kmeans.fit(df2)
hc.fit(df2)
hdb.fit(df2)
```

```python
from sklearn.metrics import silhouette_score
y_Km = kmeans.predict(df2)
y_hc = hc.fit_predict(df2)
y_hdb = hdb.fit_predict(df2)
print("KMeans Silhouette Score: ", silhouette_score(df2, kmeans.labels_))
print("Hierarchical Clustering Silhouette Score: ", silhouette_score(df2, hc.labels_))
print("HDBSCAN Silhouette Score: ", silhouette_score(df2, hdb.labels_))
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pca_df2 = pd.DataFrame(pca.fit_transform(df2))
```

```python
pca_df2
```

```python
pca.explained_variance_ratio_
```

```python
plt.step(list(range(1,10)), np.cumsum(pca.explained_variance_ratio_))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.ylabel("Variance Covered")
plt.title("Variance Covered")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(pca_df2)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.xticks(range(1,11))
plt.xlabel("number of clusters")
plt.ylabel("WCSS")
plt.show()
```
